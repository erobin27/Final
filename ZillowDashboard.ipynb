## Imports

```python
import pandas as pd
import numpy as np
import cufflinks as cf
import chart_studio.plotly as py
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
%matplotlib inline

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
cf.go_offline()

import datetime #to parse dates
from  geopy.geocoders import Nominatim
import kaleido
import pathlib
```

## Extract

**import data into pandas dataframe** </br>
The data from *Zillow* is stored on github as a *.csv* file it is then imported by URL into *pandas*

```python
url = 'https://raw.githubusercontent.com/erobin27/ZillowDashboard/master/rawdata/ListAndSalePrice.csv'
df = pd.read_csv(url,index_col=0)
df.head()

df_noNaN = df.fillna(0)
df_noNaN
```

#### List Of Dates: datelist

```python
#List of all dates
datelist = []

for col in df.columns:
    try:
        datetime.datetime.strptime(col, '%Y-%m-%d')
        datelist.append(col)
    except:
        pass
    
datelist
```

#### List of States: statelist

```python
#List of all states
statelist = df['StateName'].unique()
statelist
```

#### List of Regions: regionList

```python
#List of all states
regionList = df['RegionName'].unique()
regionList
```

## User Setup

**User Date Selection**

```python
ReportDate = '2018-08-31'

print('Dates to select from: ')
datelist
```

```python
ReportRegion = 'Raleigh, NC'

ReportPath = str(pathlib.Path().resolve())
ReportState = ReportRegion.split(',')[-1].strip()
print('Regions to select from: ')
regionList

```

## Transform

#### States by Date:    df_StateByDate

```python
newCols = df.StateName.tolist()
newCols.insert(0, 'Dates')

df_StateByDate = pd.DataFrame()
df_StateByDate['Dates'] = datelist

for index, row in df.iterrows():
    temp = []
    RegionName = row['RegionName']
    for date in datelist:
        temp.append(row[date])

    df_StateByDate[RegionName] = temp

df_StateByDate.fillna(0, inplace=True)
df_StateByDate
```

#### Date by States:    df_DateByState

```python
df_DateByState = df_StateByDate.transpose()
df_DateByState.columns = df_DateByState.iloc[0]
df_DateByState.rename({'Dates':'RegionName'}, axis='columns')
df_DateByState.drop('Dates',axis=0,inplace=True)
df_DateByState.reset_index(inplace=True)
df_DateByState.rename(columns={'index':'RegionName'}, inplace=True)
df_DateByState
```

#### Latitude Longitude: LatLon_df

```python
LatLon_df = df.copy()
LatLon_df.drop('RegionType', axis=1, inplace=True)
# for date in datelist:
#     LatLon_df.drop(date, axis=1, inplace=True)

tLon = []
tLat = []
for index, row in df.iterrows():
    RegionName = row['RegionName']
    geolocator = Nominatim(user_agent="Elijah")
    try:
        location = geolocator.geocode(RegionName)
        tLon.append(location.longitude)
        tLat.append(location.latitude)
    except:
        tLon.append(0)
        tLat.append(0)

LatLon_df['Latitude'] = tLat
LatLon_df['Longitude'] = tLon

LatLon_df
```

#### Top 10 expensive cities per month: Expensive_df

```python
Expensive_df = pd.DataFrame()
for date in datelist:
    df_t = df.nlargest(10,date)
    Expensive_df[date] = df_t['RegionName'].tolist()
Expensive_df
```

#### Percentage of Total Median Sales: PercentPerState_df   &   PercentPerCity_df

*What percentage does the state make up of the entire country for median sales price?*

```python
PercentPerState_df = pd.DataFrame(columns=['StateName'] + datelist)
PercentPerState_df['StateName'] = statelist

for date in datelist:
    percentPerDate = []
    for state in statelist:
        stateSum = df_noNaN.loc[df['StateName']==state][date].sum()
        totalSum = df_noNaN[date].sum()
        percentPerDate.append(stateSum/totalSum) #sum of the states medians
    PercentPerState_df[date] = percentPerDate

# for state in statelist:
#     percentPerDate = []
#     percentPerDatePerRegion = []
#     for date in datelist:
#         #PER STATE
#         stateSum = df_noNaN.loc[df['StateName']==state][date].sum()
#         totalSum = df[date].sum()
#         percentPerDate.append(stateSum/totalSum) #sum of the states medians
#     PercentPerState_df[state] = percentPerDate

PercentPerState_df
```

*What percentage does the city make up of the state it is in for median sales price?*

```python
PercentPerCity_df = pd.DataFrame(columns=['RegionName','StateName'] + datelist)
PercentPerCity_df['RegionName'] = df_noNaN.loc[df_noNaN['StateName']==ReportState]['RegionName'].tolist()
PercentPerCity_df['StateName'] = df_noNaN.loc[df_noNaN['StateName']==ReportState]['StateName'].tolist()

regionsInState = df_noNaN.loc[df_noNaN['StateName']==ReportState]['RegionName'].tolist()

for date in datelist:
    percentPerDate = []
    for region in regionsInState:
        regionAmount = df_noNaN.loc[df['RegionName']==region][date].sum()
        stateSum = df_noNaN.loc[df['StateName']==ReportState][date].sum()
        percentPerDate.append(regionAmount/stateSum) #sum of the states medians
    PercentPerCity_df[date] = percentPerDate

PercentPerCity_df
```

### Generate Report

```python
from IPython.display import display, HTML
graphs = [
    [ReportPath + '/figures/heatmap.html', ReportPath + '/figures/heatmap.png'],
    [ReportPath + '/figures/linechart.html', ReportPath + '/figures/linechart.png'],
    [ReportPath + '/figures/piechart.html', ReportPath + '/figures/piechart.png'],
    [ReportPath + '/figures/piechart2.html', ReportPath + '/figures/piechart2.png'],
    [ReportPath + '/figures/boxplot.html', ReportPath + '/figures/boxplot.png'],
]
    # 'https://plotly.com/~christopherp/308',
    # 'https://plotly.com/~christopherp/306',
    # 'https://plotly.com/~christopherp/300',
    # 'https://plotly.com/~christopherp/296'


def report_block_template(report_type, graph_url, caption=''):
    graph_block = '<H3>' + '{caption}' + '</H3>'

    if report_type == 'interactive':
        graph_block += '<iframe style="border: none;" src="{graph_url[1]}" width="100%" height="600px"></iframe>'
    elif report_type == 'static':
        graph_block += (''
            '<a href="{graph_url[1]}" target="_blank">' # Open the interactive graph when you click on the image
                '<img style="height: 400px;" src="{graph_url[1]}">'
            '</a>')

    report_block = ('' +
        graph_block +
        '<br>'      + # Line break
        '<a href="{graph_url[0]}" style="color: rgb(190,190,190); text-decoration: none; font-weight: 200;" target="_blank">'+
            'Click to comment and see the interactive graph' + # Direct readers to Plotly for commenting, interactive graph
        '</a>' +
        '<br>' +
        '<hr>') # horizontal line                       

    return report_block.format(graph_url=graph_url, caption=caption)


interactive_report = '<H1> Zillow Housing Sales Report for ' + ReportDate + '</H1><H2>' + ReportRegion + '</H2></br>'
static_report = ''

#ADD CHARTS HERE
interactive_report += report_block_template('interactive', graphs[0], caption='HeatMap of Median listed Sales Price in the USA on ' + ReportDate)
interactive_report += report_block_template('interactive', graphs[1], caption='Median listed sales price per month in ' + ReportRegion)
interactive_report += report_block_template('interactive', graphs[2], caption='Percentage per State of median sales price to total of all states on ' + ReportDate)
interactive_report += report_block_template('interactive', graphs[3], caption='Percentage per City of median sales price to total of the state on ' + ReportDate)
interactive_report += report_block_template('interactive', graphs[4], caption='Boxplot of house listing prices in ' + ReportRegion)

with open('report.html', 'w', encoding = 'utf8') as f:
    f.write(interactive_report)
```
